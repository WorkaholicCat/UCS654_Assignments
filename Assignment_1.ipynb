```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
data = pd.read_csv('/content/data.csv', encoding = 'unicode_escape')
```

```python
r = 102303941
a = 0.05 * (r % 7)
b = 0.3 * (r % 3 + 1)
x = data['no2']

z = x + a * np.sin(b * x)
```

```python
m = np.mean(z)
s = np.std(z)
l = 1 / (2 * (s**2))

c = 1 / (s * np.sqrt(2 * np.pi))
```
